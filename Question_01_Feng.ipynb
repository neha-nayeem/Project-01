```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1="Data Output/Toronto_Crimes_Occurred_2014_2019.csv"
file2="Data Output/Toronto_Crimes_2014.csv"
file3="Data Output/Toronto_Crimes_2015.csv"
file4="Data Output/Toronto_Crimes_2016.csv"
file5="Data Output/Toronto_Crimes_2017.csv"
file6="Data Output/Toronto_Crimes_2018.csv"
file7="Data Output/Toronto_Crimes_2019.csv"
crime_2014_2019=pd.read_csv(file1)
crime_2014=pd.read_csv(file2)
crime_2015=pd.read_csv(file3)
crime_2016=pd.read_csv(file4)
crime_2017=pd.read_csv(file5)
crime_2018=pd.read_csv(file6)
crime_2019=pd.read_csv(file7)
```

```python
# Define a function to get the number of cases based on MCI, and create a bar chart for the data
def get_number(df,year):
    mci=df[~df.duplicated(subset=["MCI"])]
    mci_list=sorted(mci["MCI"])
    num_crime=[]
    grouped_mci=df.groupby(["MCI"])["event_unique_id"].count()
    for crime in mci_list:
        print(f"In {year}, there are {grouped_mci[crime]} {crime} occurred.")
        num_crime.append(grouped_mci[crime])
    x=mci_list
    y=num_crime
    plt.bar(x,y,color="blue",alpha=0.75,align="center")
    plt.xticks(rotation=45)
    plt.xlabel("MCI")
    plt.ylabel("Number of crimes")
    plt.xlim(-1,len(num_crime))
    plt.ylim(0,max(num_crime)+1000)
    plt.legend(["Case Number"],loc="upper right")
    plt.title(f"Toronto crimes occurred {year}")
    plt.savefig(f"Image Output/Toronto Crime Numbers {year}.png")
    plt.show()
```

```python
# Run the function to get total case numbers from 2014 to 2019.
get_number(crime_2014_2019,"2014-2019")
```

```python
# Optional research: Get the data of different species of crime in different years.
# Define a function that can get the number of cases of each MCI type, and use line chart to show number changes throughout the year.
def mci_through_years(df,mci_type):
    mci_df=df.loc[df["MCI"]==mci_type,:]
    years=[2014,2015,2016,2017,2018,2019]
    case_num=[]
    for year in years:
        mci_by_year=mci_df.loc[mci_df["occurrenceyear"]==year,:]
        case_num.append(len(mci_by_year["MCI"]))
        print(f"There are {len(mci_by_year['MCI'])} {mci_type} occurred in Toronto in {year}.")
    x_axis=["2014","2015","2016","2017","2018","2019"]
    y_axis=case_num
    plt.plot(x_axis,y_axis,color="red")
    plt.xlabel("Year")
    plt.ylabel("Number of Cases")
    plt.xlim(-0.5,len(x_axis))
    plt.ylim(min(y_axis)*0.5,max(y_axis)+min(y_axis)*0.5)
    plt.legend(["Case Number"],loc="upper right")
    plt.title(f"{mci_type} Occurrence in Toronto from 2014 to 2019")
    plt.savefig(f"Image Output/{mci_type} Occurrence in Toronto 2014-2019.png")
    plt.show()
```

```python
# Assault in Toronto
mci_through_years(crime_2014_2019,"Assault")
```

```python
# Observation: The assault cases keep increasing in Toronto from 2014 to 2019.
```

```python
# Auto Theft in Toronto
mci_through_years(crime_2014_2019,"Auto Theft")
```

```python
# Observation: The auto theft in Toront went down from 2015 to 2016 comparing to 2014, but the number headed up since 2017.
```
